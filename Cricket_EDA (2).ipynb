```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing the dataset
df = pd.read_excel('cricket_data.xlsx')
df.head()
```

```python
#Renaming the column headers

df = df.rename(columns = {'Mat': 'Matches', 'NO':'Not Outs', 'HS': 'Highest Inns Score', 'BF':'Balls Faced',
                    'SR': 'Batting Strike Rate'})
```

```python
df.head(5)
```

```python
df =df.dropna()
```

```python
#checking the null values
df.isnull().any()
```

```python
df.describe()
```

```python
#drop duplicates

df.duplicated()
```

```python
#split up span into and End Date

df['Span'].str.split(pat='-')
```

```python
df['Span'].str.split(pat = '-').str[1]
```

```python
df['Rookie Year'] = df['Span'].str.split(pat='-').str[0]
```

```python
df['Final Year'] = df['Span'].str.split(pat='-').str[1]
```

```python
df
```

```python
#drop span columns

df = df.drop(columns=['Span'])
```

```python
df.head(5)
```

```python
#Spliting up the country from the year

df['Player'].str.split(pat = ',')
```

```python
df['Country'] = df['Player'].str.split(pat='(').str[1]
df.head()
```

```python
# Assuming the DataFrame is named 'df' and the column is named 'Player'
df['Country'] = df['Country'].str.replace(')', '', regex=False)
```

```python
df.head()
```

```python
df['Player'] = df['Player'].str.split(pat='(').str[0]
df.head()
```

```python
#Checking the data types
df.dtypes
```

```python
# Removing the symbols from the 'Hghest Inns Score' column
df['Highest Inns Score'] = df['Highest Inns Score'].str.replace('*', '', regex=False)
df.head()
```

```python
# Removing the symbols from the 'Balls Faced' column
df['Balls Faced'] = df['Balls Faced'].str.replace('+','', regex=False)
df.head()
```

```python
df['Highest Inns Score'] = df['Highest Inns Score'].astype('int')
```

```python
df.dtypes
```

```python
#Converting Rookie Year from object to an int
df['Rookie Year'] = df['Rookie Year'].astype('int')
```

```python
df.dtypes
```

```python
#Converting Final Year from object to an int

df['Final Year'] = df['Final Year'].astype('int')
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Dropping rows with missing values 'Hon.FS Jackson'
df = df.drop(54, axis = 0)
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Replace all '-' values with '0' and then convert the column to 'int' data type
df['Balls Faced'] = df['Balls Faced'].str.replace('-', '0').astype('int')
```

```python
df.dtypes
```

```python
#Carrer lenght of the players

df['Career Length'] = (df['Final Year'].astype(int) - df['Rookie Year'].astype(int))
```

```python
df
```

```python
#Average carrer length of the players

df['Career Length'].mean()
```

```python
#Average career length of Batting Strike Rate for players who played over 10 years

df[ df['Career Length']>10] ['Batting Strike Rate'].mean()
```

```python
#Players that played before 1980

df[df['Rookie Year']<1980]
```

```python
#Count of Players that played before 1980

df[df['Rookie Year']<1980]['Player'].count()
```

```python
#Maximum Highest Inns Scored by a country and sorting in ascending order
df.groupby('Country')['Highest Inns Score'].max().sort_values(ascending=False)
```

```python
#Maximum Highest Inns Scored by a country and sorting in ascending order and adding a data frame to it.

df.groupby('Country')['Highest Inns Score'].max()\
  .to_frame('Highest by Country')\
  .reset_index()\
  .sort_values('Highest by Country', ascending=False)
```

```python
#Find the average of Hundreds, fifties, ducks, (0) by country

df.groupby('Country',)[['100', '50', '0']].mean()
```

```python
#Find the average scored of Matches, Inns, Not Outs, Runs, Highest Inns Score, Balls Faced by Player

df.groupby('Player',)[['Matches', 'Inns', 'Not Outs', 'Runs', 'Highest Inns Score', 'Balls Faced']].mean()
```
